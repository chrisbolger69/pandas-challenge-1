## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

# Create the column line_total
def line_subtotal(row):
    return row['unit_price']*row['qty']

# Add the column to the DataFrame
df['line_subtotal'] = df.apply(line_subtotal, axis = 1)

# Display the column with some other columns from the DataFrame.
df[['unit_price','qty','line_subtotal']].head(2)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

# Define a column for total weight which is the unit_weight multiplied by the qty.
def total_weight(row):
    return row['unit_weight']*row['qty']

# Add that column to the DataFrame.
df['total_weight'] = df.apply(total_weight, axis = 1)

# Define a column called shipping price which is $10 per lb for total weights 50 lbs
# and under and $7 per lb for total weights over 50 lbs.
def shipping_price(row):
    if row['total_weight'] > 50 :
        return row['total_weight']*7
    return row['total_weight']*10

# Add that column to the DataFrame
df['shipping_price'] = df.apply(shipping_price, axis = 1)

# Display the two new columns along with some others in the DataFrame
df[['unit_price','unit_weight','qty','total_weight','shipping_price']].head(3)
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

# Create line_price column by adding the line subtotal to the shipping price,
# multiplying the sum by 1.0925 to bring the tax in.
def line_price(row):
    return (row['line_subtotal']+row['shipping_price'])*1.0925

# Add that column to the DataFrame and round to the nearest hundreth.
df['line_price'] = df.apply(line_price, axis = 1).round(2)

# Display new column and some others.
df[['line_subtotal','shipping_price','line_price']].head(3)
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

# Create line_cost column by multiplying unit cost and the quantity then adding the shipping price to it.
def line_cost(row):
    return (row['unit_cost'] * row['qty']) + row['shipping_price']

# Add that column to the DataFrame and round to the nearest hundreth.
df['line_cost'] = df.apply(line_cost, axis = 1).round(2)

# Display new column and some others.
df.head(3)
```

```python
# Create a column for the profit of
# each line using line cost and line price

# Create line_profit column by subtracting the line cost from the line price.
def line_profit(row):
    return row['line_price'] - row['line_cost']
    
# Add that column to the DataFrame and round to the nearest hundreth.
df['line_profit'] = df.apply(line_profit, axis = 1).round(2)

# Display new column and some others.
df.head(3)
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

# Grab summary data for the clients on quantity and put them in variables.
client_qty_1 = df['qty'].loc[df['client_id'] == 33615].sum()
client_qty_2 = df['qty'].loc[df['client_id'] == 66037].sum()
client_qty_3 = df['qty'].loc[df['client_id'] == 46820].sum()
client_qty_4 = df['qty'].loc[df['client_id'] == 38378].sum()
client_qty_5 = df['qty'].loc[df['client_id'] == 24741].sum()

# Grab summary data for the clients on shipping price and put them in variables.
client_sp_1 = df['shipping_price'].loc[df['client_id'] == 33615].sum()
client_sp_2 = df['shipping_price'].loc[df['client_id'] == 66037].sum()
client_sp_3 = df['shipping_price'].loc[df['client_id'] == 46820].sum()
client_sp_4 = df['shipping_price'].loc[df['client_id'] == 38378].sum()
client_sp_5 = df['shipping_price'].loc[df['client_id'] == 24741].sum()

# Grab summary data for the clients on line price and put them in variables.
client_lp_1 = df['line_price'].loc[df['client_id'] == 33615].sum()
client_lp_2 = df['line_price'].loc[df['client_id'] == 66037].sum()
client_lp_3 = df['line_price'].loc[df['client_id'] == 46820].sum()
client_lp_4 = df['line_price'].loc[df['client_id'] == 38378].sum()
client_lp_5 = df['line_price'].loc[df['client_id'] == 24741].sum()

# Grab summary data for the clients on line cost and put them in variables.
client_lc_1 = df['line_cost'].loc[df['client_id'] == 33615].sum()
client_lc_2 = df['line_cost'].loc[df['client_id'] == 66037].sum()
client_lc_3 = df['line_cost'].loc[df['client_id'] == 46820].sum()
client_lc_4 = df['line_cost'].loc[df['client_id'] == 38378].sum()
client_lc_5 = df['line_cost'].loc[df['client_id'] == 24741].sum()

# Grab summary data for the clients on line profit and put them in variables.
client_lprof_1 = df['line_profit'].loc[df['client_id'] == 33615].sum()
client_lprof_2 = df['line_profit'].loc[df['client_id'] == 66037].sum()
client_lprof_3 = df['line_profit'].loc[df['client_id'] == 46820].sum()
client_lprof_4 = df['line_profit'].loc[df['client_id'] == 38378].sum()
client_lprof_5 = df['line_profit'].loc[df['client_id'] == 24741].sum()

# Create the new summary DataFrame.
client_info_df = pd.DataFrame({
    "client_id":[33615, 66037, 46820, 38378, 24741],
    "qty":[client_qty_1,client_qty_2,client_qty_3,client_qty_4,client_qty_5],
    "shipping_price":[client_sp_1,client_sp_2,client_sp_3,client_sp_4,client_sp_5],
    "line_price":[client_lp_1,client_lp_2,client_lp_3,client_lp_4,client_lp_5],
    "line_cost":[client_lc_1,client_lc_2,client_lc_3,client_lc_4,client_lc_5],
    "line_profit":[client_lprof_1,client_lprof_2,client_lprof_3,client_lprof_4,client_lprof_5]
    })

# Sort the DataFrame by total profit, and create a new dataframe.
client_info_sorted_df = client_info_df.sort_values("line_profit", ascending=False)
client_info_sorted_df
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

# Renaming the columns
client_info_renamed_df = client_info_sorted_df.rename(columns = {
"client_id":"Client ID",	
"qty":"Units",	
"shipping_price":"Shipping",
"line_price":"Total Revenue",
"line_cost":"Total Cost",	
"line_profit":"Total Profit"
})

#Rescale the dollar values into millions for shipping.
def scale_to_millions(row):
    return f"${row['Shipping'] / 1000000:.2f}M"

client_info_renamed_df['Shipping'] = client_info_renamed_df.apply(scale_to_millions, axis = 1)

#Rescale the dollar values into millions for total revenue.
def scale_to_millions(row):
    return f"${row['Total Revenue'] / 1000000:.2f}M"

client_info_renamed_df['Total Revenue'] = client_info_renamed_df.apply(scale_to_millions, axis = 1)

#Rescale the dollar values into millions for total cost.
def scale_to_millions(row):
    return f"${row['Total Cost'] / 1000000:.2f}M"

client_info_renamed_df['Total Cost'] = client_info_renamed_df.apply(scale_to_millions, axis = 1)

#Rescale the dollar values into millions for total profit.
def scale_to_millions(row):
    return f"${row['Total Profit'] / 1000000:.2f}M"

client_info_renamed_df['Total Profit'] = client_info_renamed_df.apply(scale_to_millions, axis = 1)
client_info_renamed_df.head()
```
